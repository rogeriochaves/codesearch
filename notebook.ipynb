```python
def extract_code(text):
    result = nlp(text)
    
    current_code = ""
    code_blocks = []
    for ner in result:
        entity = ner['entity']
        if 'Code_Block' not in entity and 'Function' not in entity:
            continue

        word = ner['word'].replace("Ġ", " ")
        if entity.startswith('B-'):
            if current_code != "":
                code_blocks.append(current_code.strip())
            current_code = word.strip()
        else:
            current_code += word
    if current_code != "":
        code_blocks.append(current_code)
    code_blocks = list(set(code_blocks))
    
    return code_blocks

extract_code(sequence)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.google.com/search?hl=en&q=bash replace text in a file'
headers = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'Cookie': '123',
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.16; rv:86.0) Gecko/20100101 Firefox/86.0',
    'Host': 'www.google.com',
    'Referer': 'https://www.google.com/'
}

req = requests.get(url, headers=headers)
soup = BeautifulSoup(req.text, features='html.parser')
```

```python
search_results = soup.find(id="search")

search_text = " ".join(search_results.strings).replace(u'\xa0', u' ')

%time extract_code(search_text)
```

```python
search_results = soup.find(id="search")
search_results = search_results.find_all(class_="g")
descriptions = []
for result in search_results:
    description = result.find(lambda tag: len(tag.text) > 140)
    if description:
        descriptions.append(" ".join(description.strings))

descriptions = "\n\n".join(descriptions).replace(u'\xa0', u' ')

%time extract_code(descriptions)
```

```python
import urllib

def search_and_extract_code(query):
    url = 'https://www.google.com/search?hl=en&' + urllib.parse.urlencode({"q": query})
    headers = {
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
        'Cookie': '123',
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.16; rv:86.0) Gecko/20100101 Firefox/86.0',
        'Host': 'www.google.com',
        'Referer': 'https://www.google.com/'
    }

    req = requests.get(url, headers=headers)
    soup = BeautifulSoup(req.text, features='html.parser')
    search_results = soup.find(id="search")
    search_text = "".join(search_results.strings).replace(u'\xa0', u' ')

    print(search_text)
    return extract_code(search_text)

search_and_extract_code("python urlencode")
```
